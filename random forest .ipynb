```python
import numpy as np
from sklearn import neighbors
from sklearn import metrics
from sklearn.utils import shuffle
from sklearn import ensemble
import matplotlib.pyplot as plt 


"""
experimentation:
min_leaf_samples

"""


X = np.genfromtxt('data/X_train.txt', delimiter=None)
Y = np.genfromtxt('data/Y_train.txt', delimiter=None)

X,Y = shuffle(X, Y)

Xtr = X[:10000,:]
Ytr = Y[:10000]

Xval = X[10000:20000]
Yval = Y[10000:20000]
r=[]
t=[]
v=[]
l=[1,2,4,5,6,8,10,12,16,20]
for i in l:
    rfc = ensemble.RandomForestClassifier(
        n_estimators=500, min_samples_leaf=i, n_jobs=-1, oob_score=True)

    rfc.fit(Xtr, Ytr)

    rfcRoc = metrics.roc_auc_score(Yval, rfc.predict_proba(Xval)[:,1])

    r.append( rfcRoc)
    t.append( 1 - rfc.score(Xtr, Ytr))
    v.append(1 - rfc.score(Xval, Yval))
plt.semilogx(l, t, 'r-', lw=4,alpha=0.75, label='training')
plt.semilogx(l, v, 'g-', lw=4,alpha=0.75, label='Validation')
plt.semilogx(l, r, 'b-', lw=4,alpha=0.75, label='ROC')
plt.legend()
plt.show()
```
